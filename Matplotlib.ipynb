### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Line Chart
# On the bases of the time

years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
sales = [150, 200, 250, 300, 100, 60, 80, 90, 60, 60]

# Create a line chart
plt.plot(years, sales, marker='o')
plt.title('Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(years, rotation=45)
plt.tight_layout()

plt.show()
```
