### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Line Chart
# On the bases of the time

years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
sales = [150, 200, 250, 300, 100, 60, 80, 90, 60, 60]

# Create a line chart
plt.plot(years, sales, marker='o')
plt.title('Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(years, rotation=45)
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
start_date = datetime(2024,1,1)
end_date = datetime(2024,12,31)

num_days = (end_date - start_date).days
num_days
```

```python
date_list=[]
for i in range(num_days):
    date_list.append(start_date.date())
    start_date += timedelta(days=1)
    
# date_list
```

```python
np.random.seed(0)  # For reproducibility
stock_price_list1=(np.random.randn(num_days).cumsum()*40).tolist()
stock_price_list2=(np.random.randn(num_days).cumsum()*10).tolist()
stock_price_list1
```

```python
plt.plot(date_list, stock_price_list1, color="red", linestyle="--", )
plt.plot(date_list, stock_price_list2, color="green", linestyle="--", )
plt.title('Stock Price Over Days')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.grid(True)
plt.legend(['Stock Price'])
# plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
